```python
import qrcode


PRODUCTS = []
Exit = False

#match is like switch in python
def choice(ch):
    match ch:
        case 1:
            Add()
        case 2:
            Edit()
        case 3:
            Remove()
        case 4:
            Search()
        case 5:
            Show_list()
        case 6:
            Buy()
        case 7:
            code = input("please enter your code: ")
            generate_qr_code(code)
        case 8:
            write_to_database()
            save_to_file()
            
        case _:
            print("Your choice is not in the list, please choose again")
            Show_menu()

#writed as Set
def Show_menu():
    menu_options = {
        1: "Add",
        2: "Edit",
        3: "Remove",
        4: "Search",
        5: "Show Menu",
        6: "Buy",
        7: "Generate QR Code",
        8: "Exit",
    }
    for option, description in menu_options.items():
        print(f"{option} -> {description}")

def read_from_database():
    with open('7/database.txt', 'r') as file:
        PRODUCTS.extend(
            {
                "code": data[0],
                "name": data[1],
                "price": data[2],
                "count": data[3].strip()  # Remove newline character
            } for data in (line.split(",") for line in file) #nested generator expression  
        )


def Add():
    keys = ['code', 'name', 'price', 'count']
    new_product = {key: input(f"Enter {key}: ") for key in keys} #dictionary comprehension
    PRODUCTS.append(new_product)


def Edit():
    code = input("Enter the product code: ")
    for product in PRODUCTS:
        if product["code"] == code:
            fields = {1: "name", 2: "price", 3: "count"}
            print("Select the field to edit:")
            for choice, field in fields.items():
                print(f"{choice}- {field.capitalize()}")
            choice = int(input("Enter your section to Edit(Section Num): "))
            if choice in fields:
                new_value = input(f"Enter the new {fields[choice]}: ")
                product[fields[choice]] = new_value
                print("Information updated successfully.")
            else:
                print("Invalid choice.")
            return
    print("Product not found.")


def Remove():
    code = input("Enter the product code to remove: ")
    initial_len = len(PRODUCTS)
    PRODUCTS[:] = [product for product in PRODUCTS if product["code"] != code]
    if len(PRODUCTS) < initial_len:
        print("Product successfully removed.")
    else:
        print("Product not found.")


def generate_qr_code(code):
    for product in PRODUCTS:
        if product["code"] == code:
            file_name = product["name"]+"\n" + product["price"]+"\n"+product["count"]
            name=product["name"]
            img_QR = qrcode.make(file_name)
            type(img_QR)
            img_QR.save(name+".png")
            print("Your product QR code successfully generated8")
 
            break
    else:
        print("product not found!")


def Search():
    user_input = input("Enter user keyword: ")
    product_dict = {product["code"]: product for product in PRODUCTS}
    product_dict.update({product["name"]: product for product in PRODUCTS})
    if user_input in product_dict:
        product = product_dict[user_input]
        print("code \t\t  name  \t\t price")
        print("-" * 50)
        print(f"{product['code']}\t\t{product['name']}\t\t{product['price']}\n")
    else:
        print("Not found.")


def Show_list():
    print("code \t\t  name  \t\t price")
    print("-" * 50)
    for product in PRODUCTS:
        print(f"{product['code']}\t\t{product['name']}\t\t{product['price']}")


def Buy():
    while True:
        code = input("Enter the product code: ")
        product_dict = {product["code"]: product for product in PRODUCTS}
        if code in product_dict:
            product = product_dict[code]
            quantity = int(input("Enter quantity: "))
            if int(product["count"]) >= quantity:
                product["count"] = str(int(product["count"]) - quantity)
                print("Product added to cart.")
            else:
                print("Insufficient stock.")
        else:
            print("Product not found.")
        choice = input("Do you want to buy more products? (y/n): ")
        if choice.lower() != 'y':
            break


def write_to_database():
    with open('7/database.txt', 'w') as f:
        for product in PRODUCTS:
            f.write(f"{product['code']},{product['name']},{product['price']},{product['count']}\n")

def save_to_file():
    with open('7/data_backup.txt', 'w') as f:
        for product in PRODUCTS:
            f.write(str(product) + "\n")

print('Welcome to 🛒')
read_from_database()

while True:
    Show_menu()
    UserChoice = int(input('Enter your choice: '))
    choice(UserChoice)
    if UserChoice ==8: break
```
