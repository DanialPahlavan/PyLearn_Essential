```python
import qrcode

ITEMS = []

def user_selection(ch):
    if ch == 1:
        add_item()
    elif ch == 2:
        modify_item()
    elif ch == 3:
        delete_item()
    elif ch == 4:
        find_item()
    elif ch == 5:
        display_items()
    elif ch == 6:
        purchase_item()
    elif ch == 7:
        code = input("Please enter the item code: ")
        create_qr_code(code)
    elif ch == 8:
        write_to_database()
        save_to_file()
        exit(0)
    else:
        print("Invalid selection. Please select again.")
        display_menu()

def display_menu():
    print('1 -> Add Item')
    print('2 -> Modify Item')
    print('3 -> Delete Item')
    print('4 -> Find Item by ID')
    print('5 -> Display Item List')
    print('6 -> Purchase')
    print('7 -> Create QRCode[Item]')
    print('8 -> Exit')


def load_from_database():
    with open('7/database.txt', 'r') as f:
        for line in f:
            result = line.split(",")
            item_info = {
                "code": result[0],
                "name": result[1],
                "price": result[2],
                "count": result[3].strip()  # Remove newline character
            }
            ITEMS.append(item_info)

def add_item():
    code = input("Enter item code: ")
    name = input("Enter item name: ")
    price = input("Enter item price: ")
    count = input("Enter item count: ")
    new_item = {'code': code, 'name': name, 'price': price, 'count': count}
    ITEMS.append(new_item)

def modify_item():
    code = input("Enter the item code: ")
    for item in ITEMS:
        if item["code"] == code:
            print("Select the field to modify:")
            print("1- Name")
            print("2- Price")
            print("3- Count")
            user_selection = int(input("Enter your selection: "))
            if user_selection == 1:
                new_name = input("Enter the new name: ")
                item["name"] = new_name
            elif user_selection == 2:
                new_price = input("Enter the new price: ")
                item["price"] = new_price
            elif user_selection == 3:
                new_count = input("Enter the new count: ")
                item["count"] = new_count
            else:
                print("Invalid selection.")
                return
            print("Information updated successfully.")
            return
    print("Item not found.")

def delete_item():
    code = input("Enter the item code to delete: ")
    for item in ITEMS:
        if item["code"] == code:
            ITEMS.remove(item)
            print("Item successfully deleted.")
            return
    print("Item not found.")

def create_qr_code(code):
    for item in ITEMS:
        if item["code"] == code:
            file_name = item["name"] + "\n" + item["price"] + "\n" + item["count"]
            img_qr = qrcode.make(file_name)
            img_qr.save(item["name"] + ".png")
            print("Your item QR code has been successfully created.")
            break
    else:
        print("Item not found!")


def find_item():
    user_input = input("Enter user keyword: ")
    for item in ITEMS:
        if item["code"] == user_input or item["name"] == user_input:
            print("Code\t\tName\t\tPrice")
            print("-" * 50)
            print(f"{item['code']}\t\t{item['name']}\t\t{item['price']}\t\t")
            break
    else:
        print("Not found.")

def display_items():
    print("Code\t\tName\t\tPrice")
    print("-" * 50)
    for item in ITEMS:
        print(f"{item['code']}\t\t{item['name']}\t\t{item['price']}\t\t")

def purchase_item():
    while True:
        code = input("Enter the item code: ")
        for item in ITEMS:
            if item["code"] == code:
                if int(item["count"]) > 0:
                    item["count"] = str(int(item["count"]) - 1)
                    print("Item successfully purchased.")
                    return
                else:
                    print("Item out of stock.")
                    return
        print("Item not found.")

def write_to_database():
    with open('7/database.txt', 'w') as f:
        for product in display_items():
            f.write(f"{product['code']},{product['name']},{product['price']},{product['count']}\n")

def save_to_file():
    with open('7/data_backup.txt', 'w') as f:
        for product in display_items():
            f.write(str(product) + "\n")

print('Welcome to Py🛒')
load_from_database()
print('Loading... Data is loaded successfully.')

while True:
    display_menu()
    user_selection(int(input('Enter your choice: ')))
```
